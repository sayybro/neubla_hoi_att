```python
from PIL import Image
import matplotlib.pyplot as plt
import requests
def plot_results(file_name, gt, boxes):
    req = requests.get(file_name,stream=True).raw
    pil_img = Image.open(req)
    print('get image')
    plt.figure(figsize=(16,10))
    plt.imshow(pil_img)
    ax = plt.gca()
    COLORS = [[0.000, 0.447, 0.741], [0.850, 0.325, 0.098], [0.929, 0.694, 0.125],
          [0.494, 0.184, 0.556], [0.466, 0.674, 0.188], [0.301, 0.745, 0.933]]
    colors = COLORS * 100
    for p, (xmin, ymin, xmax, ymax), c in zip(gt, boxes, colors):
        ax.add_patch(plt.Rectangle((xmin, ymin), xmax-xmin, ymax-ymin,
                                   fill=False, color=c, linewidth=3))
        cls_name = [attribute_index.keys()[attribute_index.values().index(i)] for i in p]
        cls = '/'
        
        text = f'{cls}'
        ax.text(xmin, ymin, text, fontsize=15,
                bbox=dict(facecolor='yellow', alpha=0.5))
    plt.axis('off')
#     plt.savefig('./vis/vis.jpg')
```

```python
plot_results(train[0]['file_name'],train[0]['pos_att_id'],train[0]['boxes'])
```
